```python
import dataclasses

@dataclasses.dataclass
class multivariate_normal_scratch:
    r"""
    Description

    Methods
    -------
    pdf

    
    References
    ----------
    """
    mean: np.array
    cov: np.array
    def pdf(self,X):
        "Probability density function"
        result=list()
        p = len(self.mean)

        if X.ndim>1:
            for x in X:
                pdf = ((1/(2*np.pi)**(p/2))*(np.linalg.det(self.cov)**0.5))*math.exp(-0.5*(x-self.mean).T@np.linalg.inv(self.cov)@(x-self.mean))
                result.append(pdf)
            return result
        else:
            pdf = ((1/(2*np.pi)**(p/2))*(np.linalg.det(self.cov)**0.5))*math.exp(-0.5*(X-self.mean).T@np.linalg.inv(self.cov)@(X-self.mean))
            return pdf
            
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy.random import multivariate_normal
import math
```

```python
def create_cov(sigma_list):
    """
    input:sigma_1^2,sigma_2^2,,sigma_12
    output:
    sigma_1^2,sigma_12
    sigma_12,sigma_2^2
    """
    S=np.array([[sigma_list[0],sigma_list[2]],
             [sigma_list[2],sigma_list[1]]])
    return S
```

```python
m=np.array([0,0])
S_list=[[1,1,0],
       [0.2,0.2,0],
       [2,2,0],
       [0.2,2,0],
       [2,0.2,0],
       [1,1,0.5],
       [0.3,2,0.5],
       [0.3,2,-0.5]
    
       ]



```

```python
plt.style.use('ggplot')
fig,axs=plt.subplots(4,2,figsize=(10,20),sharex=True,sharey=True)
axs=axs.ravel()
i=0
for S in S_list:
    cov=create_cov(S)
    sample=multivariate_normal(m,cov,size=500)
   # print(sample[0])
    axs[i].set_title(f'{S}')
    sns.scatterplot(x=sample[:,0],y=sample[:,1],ax=axs[i])
    i+=1
```
